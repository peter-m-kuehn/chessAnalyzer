```python
import chess

def piece_value(piece):
        values = {chess.PAWN: 1, chess.KNIGHT: 3, chess.BISHOP: 3, chess.ROOK: 5, chess.QUEEN: 9, chess.KING: 1000}
        val = values.get(piece.piece_type, 0)
        return val

board = chess.Board()

fen = "r2qkb1r/pp1n1ppp/2p1bn2/1B1p4/3Pp3/P3P3/1PPN1PPP/R1BQK1NR w KQkq - 0 9"
board = chess.Board(fen)
print (board)

move = chess.Move.from_uci('b5c6')
is_capture = board.is_capture(move)
captured_piece = board.piece_at(move.to_square)
moved_piece = board.piece_at(move.from_square)
if is_capture and piece_value(moved_piece) > piece_value(captured_piece):
    is_sacrifice = True
else:
    is_sacrifice = False
is_check = board.is_check()
is_game_over = board.is_game_over()

print()
print("is capture: ", is_capture)
print("captured piece: ", captured_piece)
print("moved piece: ", moved_piece)
print("is sacrifice: ", is_sacrifice) 
print("is check: ", is_check) 
print("is game over: ", is_game_over)
print()
print (board)
```
