```python
import chess
import chess.engine
import chess.pgn
from IPython.display import Image, display

# performance tuning parameters
num_hreads=14 #20
hash_mb=24000 #48000

# analysis limits
max_depth = None
movetime_sec = 3
num_moves_to_return = 3

#chess.engine.Mate(m).score(mate\_score=1800)
engine_path = "/home/peterk/stockfish/stockfish-ubuntu-x86-64-avx2"
engine = chess.engine.SimpleEngine.popen_uci(engine_path)
engine.configure({'Threads':num_hreads, "Hash": hash_mb})

def analyze_position(fen):
    search_limit = chess.engine.Limit(depth=max_depth, time=movetime_sec)
    board = chess.Board(fen)
    infos = engine.analyse(board, search_limit, multipv=num_moves_to_return) #  , game=object()
    print (infos)
    print ("----------------------")
    return [format_info(info) for info in infos]


def format_info(info):
    # Normalize by always looking from White's perspective
    score = info["score"].white()

    # Split up the score into a mate score and a centipawn score
    mscore = score.mate()
    if mscore != None: centipawn_score = chess.engine.Mate(mscore).score(mate_score=100000)
    else: centipawn_score = score.score()
    return {
        "mate_score": mscore,
        "centipawn_score": centipawn_score,
        "pv": format_moves(info["pv"]),
    }


# Convert the move class to a standard string
def format_moves(pv):
    return [move.uci() for move in pv[0:6]]


with open("../data/opera_game.pgn", encoding="utf8") as pgn:
    game = chess.pgn.read_game(pgn)
    while game != None:
        board = game.board()
        display(board)
        if not board.outcome():
            for line in analyze_position(
                board.fen()
            ):
                print(line)
        game = game.next()
    engine.quit()
        
```
